```python
brain_routes=glob.glob('brain_routing/*')
age_routes=glob.glob('age_routing/*')
```

```python
for df in brain_routes:
    df=pickle.load(open(df,'rb'))
    sns.set(font_scale=0.8)
    plt.figure(figsize=(30,30))
    sns.heatmap((df-df.mean())/df.std())
    plt.xlabel('Capsules')
    plt.ylabel('Brain Cancers')
```
